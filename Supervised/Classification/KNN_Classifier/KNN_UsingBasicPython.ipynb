```python
#### Read data using pandas dataFrame and inspect the first 5 records.
import pandas as pd
df=pd.read_excel('./Data/t_shirt_size.xlsx')
df.head()
```

```python
#### Splitting train and test dataset
from sklearn.model_selection import train_test_split
x_train, x_test = train_test_split(df, test_size=0.3,random_state=0)
#### Reseting index for Train and Test DataSet
x_test=x_test.reset_index(drop=True)
x_train=x_train.reset_index(drop=True)
```

```python
#### Finding the squared distance between the test observation and each training set
for i in range(0,len(x_test)):
    k='distance '+str(x_test['Height (in cms)'][i])+','+str(x_test['Weight (in kgs)'][i])
    x_train[k]=(((x_train['Height (in cms)']-x_test['Height (in cms)'][i])**2)+((x_train['Weight (in kgs)']-x_test['Weight (in kgs)'][i])**2))**0.5
x_train
```

```python
#### Predicting T Shirt Size for Test DataSet
list2=[]
for i in range(0,len(x_test)):
    # 
    k='distance '+str(x_test['Height (in cms)'][i])+','+str(x_test['Weight (in kgs)'][i])
    df_1=x_train[['T Shirt Size',k]].sort_values([k],ascending=[1]).head(3)
    list2.append(df_1['T Shirt Size'].mode()[0])
x_test['predicted_t_shirt_size']=list2
x_test
```
